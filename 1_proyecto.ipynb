```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = r"C:\Users\orlan\Documents\Proyecto Machine Learning\datos_demograficos.csv"

df = pd.read_csv(ruta_archivo, delimiter=';')

df['Total'] = df['Total'].str.replace(',', '.')
df['Total'] = df['Total'].str.replace('..', '.')

df['Total'] = pd.to_numeric(df['Total'], errors='coerce')

df = df.dropna(subset=['Total'])

df['ID_Cliente'] = np.arange(1, len(df) + 1)
df['Ingresos'] = np.random.normal(30000, 10000, len(df)).clip(min=15000)
df['Compro_Seguro'] = np.random.choice([0, 1], len(df), p=[0.7, 0.3])
df['Edad'] = np.random.choice(['18-25', '26-35', '36-45', '46-55', '56-65', '65+'], len(df))
df['Estado civil'] = np.random.choice(['Soltero', 'Casado', 'Divorciado', 'Viudo'], len(df))

np.random.seed(42)
df['Tamaño_Hogar'] = np.random.choice([1, 2, 3, 4, 5], len(df))
df['Nivel_Educativo'] = np.random.choice(['Primaria', 'Secundaria', 'Universitario', 'Postgrado'], len(df))

df_final = pd.get_dummies(df, columns=['Edad', 'Sexo', 'Estado civil', 'Nivel_Educativo'], drop_first=True)

# Verificar nombres de columnas después de get_dummies
print(df_final.columns)

print(df_final.describe())

plt.figure(figsize=(10, 6))
sns.histplot(df_final['Ingresos'], bins=30, kde=False)
plt.title('Distribución de Ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.show()

edad_counts = df['Edad'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=edad_counts.index, y=edad_counts.values)
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

# Verificar el nombre correcto de la columna de género
genero_columna = [col for col in df_final.columns if 'Sexo' in col][0]

print(df_final[genero_columna].value_counts())
df_final[genero_columna].value_counts().plot(kind='bar')
plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.show()

# Verificar el nombre correcto de la columna de estado civil
estado_civil_columnas = [col for col in df_final.columns if 'Estado civil' in col]

# Generar gráficos para cada categoría de estado civil
for col in estado_civil_columnas:
    print(df_final[col].value_counts())
    df_final[col].value_counts().plot(kind='bar')
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

print(df_final['Compro_Seguro'].value_counts())
df_final['Compro_Seguro'].value_counts().plot(kind='bar')
plt.title('Distribución de Compro_Seguro')
plt.xlabel('Compro_Seguro')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Compro_Seguro', y='Ingresos', data=df_final)
plt.title('Distribución de Ingresos por Comportamiento de Compra de Seguros')
plt.xlabel('Compro Seguro')
plt.ylabel('Ingresos')
plt.show()

print("Correlación entre Ingresos y Compro_Seguro:", df_final['Ingresos'].corr(df_final['Compro_Seguro']))



```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


X = df_final.drop('Compro_Seguro', axis=1)
y = df_final['Compro_Seguro']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```
